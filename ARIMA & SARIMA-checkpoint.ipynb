## Walk Forward ARIMA

```python
df = pd.read_csv("C:/Users/Arpit/Desktop/Machine Learning/Data Files/Data Files/7. ST Academy - Time Series resource files/shampoo.csv", header=0, parse_dates=[0])
```

```python
df.head()
```

```python
train_size = int(df.shape[0]*0.7)
```

```python
train, test = df.Sales[0:train_size], df.Sales[train_size:]
```

```python
test.shape
```

```python
data = train
predict = []
for t in test:
    model = ARIMA(data, order=(5,1,0))
    model_fit = model.fit()
    y = model_fit.forecast()
    print(y)
    predict.append(y)
    data = np.append(data, t)
    data = pd.Series(data)
```

```python
predict
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test.values,predict)
mse
```

# SARIMA or Seasonal ARIMA



```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df = pd.read_csv("C:/Users/Arpit/Desktop/Machine Learning/Data Files/Data Files/7. ST Academy - Time Series resource files/us-airlines-monthly-aircraft-miles-flown.csv", header=0, parse_dates=[0])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.index = df['Month']
```

```python
result_a = seasonal_decompose(df['MilesMM'], model='multiplicative')
```

```python
result_a.plot()
```

```python
model = SARIMAX(df['MilesMM'],order=(5,1,3), seasonal_order=(1,1,1,12))
```

```python
model_fit = model.fit()
```

```python
residuals = model_fit.resid
```

```python
residuals.plot()
```

```python
output = model_fit.forecast()
```

```python
output
```

```python
model_fit.forecast(12)
```

```python
yhat = model_fit.predict()
```
