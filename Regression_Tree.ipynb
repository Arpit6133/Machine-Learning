```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Arpit/Desktop/Machine Learning/Data Files/Data Files/3. ST Academy - Decision Trees resource files/Movie_regression.csv",header = 0)
```

## Missing Value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

## Dummy Variable Creation

```python
df.head()
```

```python
df = pd.get_dummies(df,columns = ['3D_available','Genre'],drop_first = True)
```

```python
df.head()
```

## X-y split

```python
X = df.loc[:,df.columns!='Collection']
```

```python
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
y = df['Collection']
```

```python
type(y)
```

```python
y.head()
```

```python
y.shape
```

## Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

## Training Regression Tree

```python
from sklearn import tree
```

```python
regtree = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
regtree.fit(x_train,y_train)
```

## Predict values using trained model

```python
y_train_pred = regtree.predict(x_train)
y_test_pred = regtree.predict(x_test)
```

```python
y_test_pred
```

## Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```

## Plotting Decision Tree

```python
dot_data = tree.export_graphviz(regtree, out_file=None)
```

## Controlling Tree Growth

### Maximum number of levels in tree

```python
regtree1 = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
regtree1.fit(x_train,y_train)
```

```python
dot_data = tree.export_graphviz(regtree1, out_file=None,feature_names=x_train.columns,filled=True)
```
